```python
from protobuf.battle_pb2 import BattleSummary
from pprint import pprint

from analysis_models.trainer import Battle, Trainer
from typing import Tuple, List, Any

from analyzer import load_data
(battles, trainers, raw_battles) = load_data()
raw_battles: List[BattleSummary] = list(raw_battles)
battle_list: List[Battle] = list(battles.values())
trainers_list: List[Trainer] = list(trainers.values())

```

```python
from utils.battle_printer import print_battle_log

def print_battle_from_hashid(hashid: str):
    battle = [battle for battle in raw_battles if battle.seed == hashid][0]
    print_battle_log(battle)

def print_top_n_battle_for_trainer(trainer_id: Tuple, n: int):
    trainer = trainers[trainer_id]
    trainer.victories.sort(key=lambda battle: battle.losing_trainer.elo, reverse=True)
    print(f"Number {n} battle for trainer {trainer.full_name}: {trainer.victories[n].seed}")
    print_battle_from_hashid(trainer.victories[n].seed)

def print_bottom_n_battle_for_trainer(trainer_id: Tuple, n: int):
    trainer = trainers[trainer_id]
    trainer.defeats.sort(key=lambda battle: battle.winning_trainer.elo)
    print(f"Number {n} bottom battle for trainer {trainer.full_name}: {trainer.defeats[n].seed}")
    print_battle_from_hashid(trainer.defeats[n].seed)

def find_battle_between(trainer_id: Tuple, enemy_id: Tuple):
    trainer = trainers[trainer_id]
    enemy = trainers[enemy_id]
    for battle in trainer.victories + trainer.defeats:
        if battle.player == trainer and battle.enemy == enemy:
            print(f"Battle between {trainer.full_name} and {enemy.full_name}: {battle.seed}")
            print_battle_from_hashid(battle.seed)
```
